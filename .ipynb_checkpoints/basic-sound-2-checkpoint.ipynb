```python
import thinkdsp
import thinkplot
import numpy
%matplotlib inline
```

```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1, offset=0)
```

```python
cos_sig.plot();
thinkplot.config(xlabel='time(s)',legend=False)
```

```python
wave =  cos_sig.make_wave(duration=5,framerate=11025)
```

```python
wave.apodize()
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
```
